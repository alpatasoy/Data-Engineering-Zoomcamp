```python
df_iter = pd.read_csv(
    prefix + 'yellow_tripdata_2021-01.csv.gz',
    nrows=100,
    dtype=dtype,
    parse_dates=parse_dates,
    iterator= True, # activating the iterator
    chunksize = 10, # records per batch
)
```

```python
!uv add tqdm
```

```python
from tqdm.auto import tqdm
```

```python
first= True
for df_chunk in tqdm(df_iter):
    if first:
        #first one creates the empty table
        df_chunk.head(0).to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
        first = False
        print("Table created")

    df_chunk.to_sql(name='yellow_taxi_data', con=engine, if_exists='append')
    print("ingested: ", len(df_chunk))
```
